```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,5,11)

y = x ** 2

x,y
```

```python
plt.plot(x,y,)

plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

```python
## Object Oriented Method

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('BIG plot')
axes2.plot(y,x)
axes2.set_title('Small plot')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = "X Sqaured")

ax.plot(x,x**3, label = "X Cubed")

plt.tight_layout()
ax.legend(loc = 0)
```

```python

fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize =(8,2))

axes[0].plot(x,x**2)
axes[1].plot(x,x**3)

plt.tight_layout()
```
