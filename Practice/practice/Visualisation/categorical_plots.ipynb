```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

Categorical items are defined by being a binary output. This means that the output will be [Yes, No] or [0, 1]. 

An example of this is whether someone is a smoker or not. 

Barplots, countplots, and boxplots are best used to demonstrate data to CEOs and Upper management, as they are easier to understand.

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

The violinplot, stripplot, and swarmplot are a good way to explore the data and analysing it. However, it should be used to demonstrate the data to Data Scientists. 

```python
sns.violinplot(x='day', y= 'total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue = 'sex', split = True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

You can use factor plot to plot the desired plot using the kind method.

```python
sns.factorplot(x='day',y='total_bill', data=tips, kind='bar')
```
