## Terminology Used:

### Bearish

&rarr; 'Bearish Trend' in financial markets can be defined as a downward trend in the prices of an industry's stocks or the overall fall in broad market indices

### Bullish 

&rarr; "bullish" means an investor believes a stock or the overall market will go higher


In this example we are using something very simple, in which we are detecting only the engulfing patterns.

### Engulfing Candle

&rarr; Is a pattern that is compremised of two bars on a price chart. They are used to indicate a market reversal. The second candlestick will be much larger than the first candlestick, so that it completely engulfs the length of the previous bar. and a **Market Reversal** occurs when the direction of a price trend has changed, from going up to going down, or vice versa.
